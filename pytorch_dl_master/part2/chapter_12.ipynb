 + 정밀도, 재현율, 참/거짓 양성/음성 등의 정의와 계산
 + F1 점수와 그 밖의 품질 메트릭
 + 과적합 방지를 위한 데이터 밸런싱과 증강

우리는 현재 곤경에 처해 있다. 딥러닝 프로젝트가 겨우 자리 잡았지만 쓸 만한 결과는 하나도 없고 신경망은 모든 데이터를 결절이 아니라고 분류한다. 더 나쁜 상황은 훈련셋과 검증셋을 전체 분류에 대한 정확도로 평가하다 보니 결과가 표면적으로 좋게 나오는 것이다. 음성 샘플에 지나치게 치우친 데이터 덕분에 모델이 높은 점수를 얻고자 단순히 모두 음성으로 판단한다. 이런 모델은 쓸모가 없다.

이제부터, 완벽하지는 않지만 분류 모델을 어느 정도 동작하게 만들어본다. 이 장은 어떻게 측정하고 정량화하며 표현할지, 그리고 모델의 작업 수행을 어떻게 개선할지에 대해 설명한다.

## 12.1 개선을 위한 상위 계획

지금은 좁은 범위의 단일 메트릭에 너무 집중한 나머지, 모델이 일반적인 상황에서는 쓸모 없게 됐다.

재현율, 정밀도, F1점수, 밸런싱, 증강을 통해 모델이 일반적인 상황에서도 사용가능한 수준이 되게끔 한다.

## 12.2 착한 개와 나쁜 녀석 : 거짓 양성과 거짓 음성

$\textbf{거짓 양성}^{\text{false positive}}$ 이란 관심사나 원하는 부류로 분류되는 일이 발생했지만 실제로는 아닌 경우다(결절이 아닌데 결절로 분류).

$\textbf{참 양성}^{\text{true positive}}$ 이란 관심 항목을 잘 분류한 경우다(결절이 맞고 결절로 분류).

$\textbf{거짓 음성}^{\text{false negative}}$ 이란 관심사가 아니거나 관심 부류가 아니라고 판단되는 경우(결절이 맞는데 결절이 아닌걸로 분류)

$\textbf{참 음성}^{\text{true negative}}$ 관심 대상이 아닌것을 잘 분류한 경우(결절이 아닌데 결절이 아닌걸로 분류)

11장의 말미에서 전체 훈련셋과 검증셋을 평가할 때 정확도 비율에 집중했고 확실히 이 방법은 좋은 평가 방법이 아니었다.

단편적인 단일 메트릭(참 양성이나 참 음성의 수처럼)을 보면 전체 분류 성능을 나타내는 단일 메트릭이 필요하다는 생각이 든다.
