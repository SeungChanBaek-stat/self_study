### 1.3.3 변수

````python
x = 10  #변수 x에 10을 대입
print(x)    #변수 x의 값을 출력

x = 100 #변수 x의 값을 100으로 변경
print(x)    #변수 x의 값을 출력

y = 3.14    #변수 y에 3.14를 대입
print(x * y)    #x와 y를 곱한 값을 출력

type(x * y) #결과값의 자료형을 출력
````

파이썬은 $\scriptsize\textbf{동적 언어}$로 분류되는 프로그래밍 언어이다. 동적이라 함은 변수의 자료형을 상황에 맞게 자동을 결정한다는 뜻이다. 앞의 예에서 x의 자료형이 `int`(정수)라는 것을 사용자가 명시한 적이다 없다! 하지만 10이라는 정수로 초기화할 때, x의 형태가 `int`임을 파이썬이 스스로 판단하는 것이다. 또, 정수와 실수를 곱한 결과는 실수가 되었다.(자동 형변환)

### 1.3.4 리스트
여러 데이터를 $\scriptsize\textbf{fltmxm}^{\text{list}}$로도 정리할 수 있다.

````python
a = [1, 2, 3, 4, 5] #리스트를 변수 a에 대입
print(a)    #변수 a의 값을 출력
print(len(a))   #리스트의 길이를 출력
print(a[0]) #첫 원소에 접근
print(a[4]) #다섯 번째 원소에 접근
a[4] = 99   #값을 변경
print(a)    #변경된 값을 출력
````

원소에 접근할 때는 `a[0]`처럼 한다. []안의 수를 인덱스(색인)라 하며 인덱스는 0부터 시작한다.(인덱스 0이 첫 번째 원소를 가리킨다.) 또 파이썬 리스트에는 $\scriptsize\textbf{슬라이싱}^{\text{slicing}}$이라는 편리한 기법이 준비되어 있다. 슬라이싱을 이용하면 범위를 지정해 원하는 부분 리스트를 얻을 수 있다.

````python
print(a)
print(a[0:2])   #인덱스 0부터 2까지 얻기(2번째는 포함하지 않는다)
print(a[1:])    #인덱스 1부터 끝까지 얻기
print(a[:3])    #처음부터 인덱스 3까지 얻기
print(a[:-1])   #마지막 원소의 1개 앞까지 얻기
print(a[:-2])   #마지막 원소의 2개 앞까지 얻기
````

### 1.3.6 bool
파이썬에는 $\scriptsize\textbf{bool}^{\text{불 혹은 불리언}}$이라는 자료형이 있다. 이 자료형은 True(참)와 False(거짓)이라는 두 값중 하나를 취한다. 또 bool에는 and, or, not 연산자를 사용할 수 있다. 즉, 자료형에 따라 사용할 수 있는 연산자가 정해져 있는 것이다.

### 1.4.2 클래스
지금까지 `int`와 `str`등의 자료형을 살펴보았다. (그리고 `type()`함수로는 원하는 데이터의 자료형을 알아낼 수 있었다.) 이들은 내장된 자료형, 즉 파이썬이 기본으로 제공하는 자료형이다. 이번 절에서는 새로운 $\scriptsize\textbf{클래스}$를 정의한다. 개발자가 직접 클래스를 정의하면 독자적인 자료형을 만들 수 있다. 또한, 클래스에는 그 클래스만의 전용 함수(메서드)와 속성을 정의할 수도 있다.  

파이썬에서는 class라는 키워드를 사용하여 클래스를 정의한다. 클래스의 구조는 다음과 같다.

```python
class 클래스 이름:
    def __init__(self, 인수, ...):      # 생성자
        ...
    def 메서드 이름 1(self, 인수, ...):     #메서드 1
        ...
    def 메서드 이름 2(self, 인수, ...):     #메서드 2
        ...
```

클래스 정의에는 $\scriptsize\textbf{\_\_init\_\_}$이라는 특별한 메서드가 있는데, 클래스를 초기화하는 방법을 정의한다. 이 초기화용 메서드를 $\scriptsize\textbf{생성자}^{\text{constructor}}$라고도 하며, 클래스의 인스턴스가 만들어질 때 한 번만 불리게 된다. 또, 파이썬에서는 메서드의 첫 번째 인수로 자신(자신의 인스턴스)을 나타내는 self를 명시적으로 쓰는 것이 특징이다.   

그럼 간단한 클래스를 하나 연습삼아 만들어보도록 한다. 다음 코드를 man.py파일로 저장하자.

````python
class Man:
    def __init__(self, name):
        self.name = name
        print("Initialized!")
    
    def hello(self):
        print("Hello " + self.name + "!")
        
    def goodbye(self):
        print("Good-bye " + self.name + "!")
        
NAME = input()

m = Man(NAME)
m.hello()
m.goodbye()
````

여기에서는 Man이라는 새로운 클래스를 정의하였다. 그리고 Man 클래스에서 m이라는 인스턴스(객체)를 생성한다.   

Man의 생성자(초기화 메서드)는 `name`이라는 인수를 받고, 그 인수로 인스턴스 변수인 `self.name`을 초기화한다. $\scriptsize\textbf{인스턴스 변수}$는 인스턴스별로 저장되는 변수이다. 파이썬에서는 `self.name`처럼 self 다음에 속성 이름을 써서 인스턴스 변수를 작성하거나 접근할 수 있다.

### 1.5.3 넘파이의 산술 연산
다음은 넘파이 배열로 산술 연산을 수행하는 예이다.

````python
x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
print(x + y)    #원소별 덧셈
print(x - y)    #원소별 뺄셈
print(x * y)    #원소별 곱셈
print(x / y)    #원소별 나눗셈
````

여기에서 주의할 점은 배열x와 배열y의 원소 수가 같다는 것이다. (둘 다 원소를 3개씩 갖는 1차원 배열) x와 y의 원소수가 같다면 산술 연산은 각 원소에 대해서 행해진다. 원소 수가 다르면 오류가 발생하니 원소 수 맞추기는 매우 중요하다. 참고로, '원소별'이라는 말은 영어로 element-wise라고 한다. 예컨대 '원소별 곱셈'은 element-wise product라고 한다.  

넘파이 배열은 원소별 계산뿐 아니라 넘파이 배열과 수치 하나(스칼라값)의 조합으로 된 산술 연산도 수행할 수 있다. 이 경우 스칼라값과의 계산이 넘파이 배열의 원소별로 한 번씩 수행된다. 이 기능을 $\scriptsize\textbf{브로드캐스트}$라고 한다. (자세한 것은 1.5.5절에서 다시 다루기로 한다.)

````python
x = np.array([1.0, 2.0, 3.0])
print(x / 2.0)
````

### 1.5.4 넘파이의 N차원 배열
넘파이는 1차원 배열(1줄로 늘어선 배열)뿐 아니라 다차원 배열도 작성할 수 있다. 예를 들어 2차원 배열(행렬)은 다음처럼 작성한다.

````python
A = np.array([[1, 2], [3, 4]])
print(A)
print(A.shape) #행렬의 차원
print(A.dtype) #행렬에 담긴 원소의 자료형
````

방금 $2 \times 2$의 A라는 행렬을 작성했다. 행렬의 차원은 shape으로, 행렬에 담긴 원소의 자료형은 dtype으로 알 수 있다. 이어서 행렬의 산술 연산을 보도록 하자.

````python
A = np.array([[1, 2], [3, 4]])
B = np.array([[3, 0], [0, 6]])
print(A + B)    #원소별 덧셈
print(A * B)    #원소별 곱셈
````

차원이 같은 행렬끼리면 행렬의 산술 연산도 대응하는 원소별로 계산된다. 배열과 마찬가지로 말이다. 행렬과 스칼라값의 산술 연산도 가능하다. 이때도 배열과 마찬가지로 브로드캐스트 기능이 작동한다.

````python
A = np.array([[1, 2], [3, 4]])
A * 10
````

$\textbf{NOTE}$     
넘파이 배열(`np.array`)은 N차원 배열을 작성할 수 있다. 1차원 배열, 2차원 배열, 3차원 배열처럼 차수의 배열을 만들 수 있다는 뜻이다. 수학에서는 1차원 배열은 $\scriptsize \textbf{벡터}^{\text{vector}}$, 2차원 배열은 $\scriptsize \textbf{행렬}^{\text{matrix}}$이라고 부른다. 또한 벡터와 행렬을 일반화한 것을 $\scriptsize \textbf{텐서}^{\text{tensor}}$라 한다. 이 책에서는 기본적으로 2차원 배열은 '행렬', 3차원 이상의 배열을 '다차원 배열'이라 부른다.   

**참고 : 구글의 딥러닝 프레임워크인 텐서플로의 이름이 여기서 유래한다. 텐서(tensor)가 신경망을 타고 흐른다(flow)는 뜻이다. (진짜 이름 하나는 잘 짓는다)

### 1.5.6 원소접근
원소의 인덱스는 0부터 시작한다. 그리고 각 원소에 접근하려면 다음과 같이 하게 된다.

````python
X = np.array([[51, 55], [14, 19], [0, 4]])
print(X)
print(X[0]) #0행
print(X[0][1])  #(0, 1)위치의 원소
````

for 문으로도 각 원소에 접근할 수 있다.

````python
for row in X:
    print(row)
````

넘파이는 지금까지의 방법외에도, 인덱스를 배열로 지정해 한 번에 여러 원소에 접근할 수도 있다.

````python
X = X.flatten() #X를 1차원 배열로 변환(평탄화)
print(X)
print(X[np.array([0, 2, 4])]) #인덱스가 0, 2, 4인 원소 얻기
````

이 기법을 응용하면 특정 조건을 만족하는 원소만 얻을 수 있다. 예컨대 다음과 같이 배열 X에서 15 이상인 값만 구할 수 있다. (이것도 뭔가 브로드캐스팅 방식아닌가?)

````python
X > 15
````

````python
X[X > 15]
````

넘파이 배열에 부등호 연산자를 사용한 (앞 예에서 X>15) 결과는 boolean 배열이다. 역에서는 이 boolean 배열을 사용해 배열 X에서 True에 해당하는 원소, 즉 값이 15보다 큰 원소만 꺼내고 있다.

### 1.6.1 단순한 그래프 그리기

### 1.6.2 pyplot의 기능
여기에 cos 함수도 추가로 그려보도록 한다. 또, 제목과 각 축의 이름(레이블) 표시 등, $\scriptsize \textbf{pyplot}$의 다른 기능도 사용해보도록 하자.

````python
import numpy as np
import matplotlib.pyplot as plt

#데이터 준비
x = np.arange(0, 6, 0.1)    #0에서 6까지 0.1간격으로 생성
y1 = np.sin(x)
y2 = np.cos(x)

#그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")    #cos 함수는 점선으로 그리기
plt.xlabel("x")  #x축 이름
plt.ylabel("y")  #y축 이름
plt.title('sin & cos')  #제목
plt.legend()  #범례 표시
plt.show()
````

$\textbf{등장 시기 }$  
$\scriptsize \textbf{PIL (Python Imaging Library)}$: PIL은 오래된 라이브러리로, 원래 1995년경에 개발되기 시작했습니다. 이후 2011년에 PIL의 마지막 공식 버전인 1.1.7이 출시되었습니다. PIL은 그 이후로 더 이상 업데이트되지 않았으며, 현재는 Pillow라는 포크가 PIL의 후속으로 널리 사용되고 있습니다. Pillow는 PIL과 호환되면서 새로운 기능과 버그 수정을 제공합니다.   

$\scriptsize \textbf{matplotlib}$: matplotlib은 2003년경에 처음 개발되었습니다. matplotlib.image.imread() 함수는 matplotlib 라이브러리의 일부로, matplotlib가 처음 출시된 이후 어느 시점에 추가되었지만, 정확한 날짜는 PIL의 Image.open()보다는 나중일 것입니다.   

$\textbf{사용 빈도}$
$\scriptsize \textbf{PIL/Pillow}$: 이미지 처리와 조작에 있어서 Python에서 가장 기본적이고 널리 사용되는 라이브러리 중 하나입니다. Pillow는 PIL의 기능을 확장하며, 이미지 파일을 불러오고, 처리하고, 저장하는 등 다양한 기능을 제공합니다. 사용의 편리함과 광범위한 기능 덕분에 많은 개발자와 프로젝트에서 선호됩니다.  

$\scriptsize \textbf{matplotlib.image.imread()}$: 주로 데이터 시각화를 목적으로 하는 matplotlib 라이브러리의 일부입니다. 이 함수는 이미지 데이터를 불러와 시각화나 간단한 이미지 처리 목적으로 사용됩니다. matplotlib는 주로 그래프나 차트를 그리는 데 사용되지만, 이미지를 불러오는 기능도 포함하고 있습니다. 

$\textbf{결론 } $  
PIL (현재는 Pillow로 알려짐)이 matplotlib.image.imread()보다 먼저 나왔으며, 이미지 처리와 관련하여 보다 널리 사용됩니다. Pillow는 이미지 처리의 다양한 측면을 다룰 수 있는 반면, matplotlib의 imread()는 주로 이미지를 시각화 목적으로 불러오는 데 사용됩니다. 따라서, 사용자의 목적에 따라 적합한 도구를 선택할 수 있습니다.
