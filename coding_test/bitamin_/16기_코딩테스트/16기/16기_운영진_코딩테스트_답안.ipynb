### **문제 1.2 (4점)**
자연수 n이 주어졌을 때, n의 모든 약수를 찾아 합산하는 함수 sum_of_divisors(n)를 작성하세요.

*  함수 안에 부분을 채워주세요



```python
def sum_of_divisors(n):
    divisors = []

    for i in range(1, n + 1):
        if n % i == 0:
            divisors.append(i)

    total = sum(divisors)
    return total

print(sum_of_divisors(12))
```

## **문제 2.2 (5점)**

**timestamp**를 **datetime**으로 변환하고 ***year, month, hour, weekday*** 컬럼을 생성하세요.

image.png

```python
london_df['timestamp'] = pd.to_datetime(london_df['timestamp'])
london_df['year'] = london_df['timestamp'].dt.year
london_df['month'] = london_df['timestamp'].dt.month
london_df['hour'] = london_df['timestamp'].dt.hour
london_df['weekday'] = london_df['timestamp'].dt.dayofweek
london_df.head(5)
```

## **문제 2.3 (6점)**

평일,주말과 공휴일을 구분하는 새로운 변수 ***day_type (평일/주말/공휴일)*** 을 생성하시오. 이때 각각의 값은 아래 예시와 같이 **weekday,weekend,holiday**로 출력되게 하시오.

image.png

```python
def get_day_type(row):
    if row['is_holiday'] == 1.0:
        return 'Holiday'
    elif row['is_weekend'] == 1.0:
        return 'Weekend'
    else:
        return 'Weekday'

london_df['day_type'] = london_df.apply(get_day_type, axis=1)
london_df.head(5)
```

## **문제 2.4 (5점)**

***wind_speed*** 컬럼에 대해 **Min-Max 정규화**를 적용한 결과를

이후, ***wind_speed***와 ***wind_speed_norm*** 두 컬럼만 포함된
데이터프레임을 출력하세요

출력 예시:

## **문제 2.5 (8점)**

***wind_speed*** 컬럼에 대해
**IQR** 기준으로 이상치를 탐지하고 제거한 뒤,
이상치 제거 전과 후의 데이터 행 수 차이를 비교하세요.

출력 예시:
image.png

```python
def remove_outliers_column_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

wind_clean_df = remove_outliers_column_iqr(london_df, 'wind_speed')

# 전후 행 수 비교
original_rows = london_df.shape[0]
cleaned_rows = wind_clean_df.shape[0]

print(original_rows,cleaned_rows)
```

## **문제 3.1 (3점)**

총 행(row) 수와 열(column) 수를 출력하세요

```python
credit_df = pd.read_csv("UCI_Credit_Card.csv")
shape = credit_df.shape
shape
```

## **문제 3.2 (5점)**


본 데이터셋에서 신용카드 연체 여부를 예측하기 위해
***default.payment.next.month*** 컬럼을 **타겟 변수y**로 설정하고,
예측에 필요 없는 **ID 컬럼과 타겟 컬럼**을 제외한 나머지를
**입력 변수X**로 설정하세요.

그 후, 전체 데이터를 학습용 80%, 평가용 20%으로 나누세요.
단, 타겟 클래스 간의 비율이 유지되도록
**Stratified Split**을 적용해야 합니다.
**random_state**는 42로 지정해주세요

```python
X = credit_df.drop(columns=['ID', 'default.payment.next.month'])
y = credit_df['default.payment.next.month']


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42)
```

## **문제 3.3 (8점)**

**RandomForestClassifie**r를 이용해 모델을 학습한 후,
테스트셋에 대해 **정확도(Accuracy)**와 **F1-score**를 출력하세요.

```python
# 모델 학습
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# 예측 및 평가
y_pred = rf_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

accuracy, f1
```

## **문제 3.4 (9점)**

학습한 **RandomForestClassifier** 모델에서
각 피처의 중요도(feature_importances_)를 추출하세요.

그 중 중요도가 높은 상위 10개 피처를 기준으로
**막대그래프**를 시각화하세요.

색상은 **viridis**로 지정해주세요.

* **그래프 제목**: Top 10 Important Features (Random Forest)
* **x축명**: Importance
* **y축명**: Feature

image.png

```python
# 피처 중요도 추출
feature_importances = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf_model.feature_importances_
}).sort_values(by='Importance', ascending=False).head(10)

# 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=feature_importances, x='Importance', y='Feature', palette='viridis')
plt.title("Top 10 Important Features (Random Forest)")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```

## **문제 3.5 (5점)**

테스트셋에 대해 **predict_proba**로 예측 확률을 구하고,
해당 확률을 기반으로 ROC Curve의 **AUC 값**을 계산하세요.



```python
y_prob = rf_model.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

print(f"AUC: {roc_auc:.4f}")
```
